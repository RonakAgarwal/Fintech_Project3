```python
# Put all JSON results into dataframes


df_z_for_sale_meck = pd.json_normalize(data=listings_meck_json['props'])


df_z_for_sale_cab = pd.json_normalize(data=listings_csb_json['props'])


df_z_for_sale_gst = pd.json_normalize(data=listings_gst_json['props'])

df_z_for_sale_un = pd.json_normalize(data=listings_un_json['props'])


display ( df_z_for_sale_meck.head() )
display ( df_z_for_sale_cab.head() )
display ( df_z_for_sale_gst.head() )
display ( df_z_for_sale_un.head() )
```

```python
# Combine all the dataframes into one

frames = (df_z_for_sale_meck, df_z_for_sale_cab, df_z_for_sale_gst, df_z_for_sale_un)
df_z_for_sale_full = pd.concat(frames)

print('Num of rows:', len(df_z_for_sale_full))
print('Num of cols:', len(df_z_for_sale_full.columns))

df_z_for_sale_full.head()
```

```python
# List of Zillow IDs for each listing

zpid_list = df_z_for_sale_full['zpid'].tolist()
zpid_list
```

```python
# Loops through each Listing Zillow ID, calls API to get info for each one


prop_detail_list = []

for zpid in zpid_list:
  url = "https://zillow-com1.p.rapidapi.com/property"
  querystring = {"zpid":zpid}
  headers = {
      'x-rapidapi-host': "zillow-com1.p.rapidapi.com",
      'x-rapidapi-key': rapid_api_key
      }
  z_prop_detail_resp = requests.request("GET", url, headers=headers, params=querystring)
  z_prop_detail_resp_json = z_prop_detail_resp.json()

  time.sleep(1.5)

  prop_detail_list.append(z_prop_detail_resp_json)
```

```python
# Puts all results in a dataframe

df_z_prop_detail = pd.json_normalize(prop_detail_list)
print('Num of rows:', len(df_z_prop_detail))
print('Num of cols:', len(df_z_prop_detail.columns))
df_z_prop_detail.head(1)
```

```python
# Access the required columns for analysis

cols = [
 'latitude',
 'longitude',
 'streetAddress', 
 'city',
 'state',
 'zipcode',
 'livingArea',
 'price',
 'rentZestimate',
 'zpid',
 ]

# Find price to rent ratio and add it to the dataframe
df_z_prop_detail_output = df_z_prop_detail[cols]

df_z_prop_detail_output['PRR'] = df_z_prop_detail_output['price'] / (df_z_prop_detail_output ['rentZestimate'] * 12)
df_z_prop_detail_output = df_z_prop_detail_output.dropna()
df_z_prop_detail_output = df_z_prop_detail_output.sort_values(by=['PRR'])

# Sort by price to rent ratio, only keep top 10
PRR_df = df_z_prop_detail_output.head(10)
PRR_df
```

```python
# Map of the top 10 listings

scattermap = px.scatter_mapbox(
    PRR_df,
    lat='latitude',
    lon='longitude',
    hover_name="streetAddress",
    size='price',
    color='PRR',
    zoom=9,
    size_max=22,
    height= 1000
)

scattermap.update_layout(mapbox_style="dark")

scattermap.show()
```
